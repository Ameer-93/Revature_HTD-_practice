````python
"""
When you use the * operator to unpack a list and pass arguments to a function,
it’s exactly as though you’re passing every single argument alone.
This means that you can use multiple unpacking operators to get values from
several lists and pass them all to a single function.
"""

def my_sum(*args):
    result = 0
    for x in args:
        result += x
    return result

list1 = [1, 2, 3]
list2 = [4, 5]
list3 = [6, 7, 8, 9]

print(my_sum(*list1, *list2, *list3))
````

````python
my_first_list = [1, 2, 3]
my_second_list = [4, 5, 6]
my_merged_list = [*my_first_list, *my_second_list] # The unpacking operator * is prepended to both my_first_list and my_second_list.

print(my_merged_list)
````

````python
## You can even merge two different dictionaries by using the unpacking operator **:
my_first_dict = {"A": 1, "B": 2}
my_second_dict = {"C": 3, "D": 4}
my_merged_dict = {**my_first_dict, **my_second_dict}

print(my_merged_dict)
````

# Lambda Functions in Python

## 1. Introduction to Lambda Functions

Lambda functions are small, anonymous functions that can have any number of arguments but can only have one expression. They're also known as "anonymous functions" because they don't require a `def` statement with a function name.

### Key Characteristics:

- **Anonymous**: No function name required
- **Concise**: Single expression only
- **Inline**: Can be defined where they're used
- **Functional**: Often used in functional programming paradigms


### When to Use Lambda:

- Short, simple operations
- One-time use functions
- As arguments to higher-order functions
- Quick transformations in data processing

## 2. Basic Syntax and Structure

### Syntax:

```python
lambda arguments: expression
```

### Components:

- `lambda`: The keyword that defines a lambda function
- `arguments`: Parameters (can be zero or more)
- `:`: Separator between parameters and expression
- `expression`: Single expression that gets evaluated and returned

````python
def add(x, y):
    return x + y

add_lambda  = lambda x, y: x + y

print(add(2,3))
print(add_lambda(2,3))
````

````python
greet = lambda : "Hello world!"

greet()
````

````python
square = lambda x: x**2

print(square(8))

upper = lambda txt: txt.upper()

print(upper("Hello"))
````

````python
multiply = lambda a,b,c : a*b*c
multiply(2,3,4)
````

````python
# default args

power = lambda base, exp=2 :  base ** exp

print(power(10))
print(power(10, 3))
````

````python
# Conditional Expressions

max_value = lambda a, b : a if a>b else b

print(max_value(2,7))


is_even = lambda n: "Even" if n%2==0 else "Odd"

print(is_even(33))
````


### Comparison Table:

| Feature       | Lambda Function        | Regular Function            |
| ------------- | ---------------------- | --------------------------- |
| Definition    | `lambda x: x * 2`      | `def func(x): return x * 2` |
| Name          | Anonymous              | Named                       |
| Statements    | Single expression only | Multiple statements         |
| Documentation | No docstrings          | Can have docstrings         |
| Debugging     | Harder to debug        | Easier to debug             |
| Reusability   | Limited                | High                        |


````python
# Regular function - more readable for complex logic
def calculate_grade(score):
    """Calculate letter grade based on numeric score."""
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'


# Lambda - good for simple operations
grade_simple = lambda score: 'Pass' if score >= 60 else 'Fail'


print(calculate_grade(85))
print(grade_simple(85))
````

## `map()` https://realpython.com/python-map-function/
## Definition and Usage

The `map()` function executes a specified function for each item in an iterable. The item is sent to the function as a parameter.

## Syntax

map(_function_, _iterables_)

## Parameter Values

| Parameter  | Description                                                                                                                                                          |
| ---------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| _function_ | Required. The function to execute for each item                                                                                                                      |
| _iterable_ | Required. A sequence, collection or an iterator object. You can send as many iterables as you like, just make sure the function has one parameter for each iterable. |


````python
def myfunc(n):
  return len(n)

x = map(myfunc, ('apple', 'banana', 'cherry'))
````

````python
# res = map(func, iterable)
fruits = ('apple', 'banana', 'cherry')
res = map(lambda x : len(x), fruits)

list(res)
````

````python
numbers = [1, 2, 3, 4, 5]

squared = list(map(lambda x: x ** 2, numbers))
print(squared)
````

````python
# Convert temperatures from Celsius to Fahrenheit
celsius = [0, 20, 30, 40]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
````

# `filter()`

The built-in **`filter()`** function in Python is used to process an iterable and extract items that satisfy a given condition, as determined by a filtering function.

`filter()` returns an iterator that yields only the elements for which the filtering function returns a truthy value:

````python
numbers = [-2, -1, 0, 1, 2]


def is_positive(num):
    if num>0:
        return True
    else:
        False


positives = list(filter(is_positive, numbers ))
print(positives)

````

````python

positive_numbers = filter(lambda n: n > 0, numbers)
list(positive_numbers)
````

````python
# Filter even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)
````

````python
# Filter strings longer than 5 characters
words = ['cat', 'elephant', 'dog', 'hippopotamus', 'rat']
long_words = list(filter(lambda w: len(w) > 5, words))
print(long_words)
````

````python
nums = [1,2,3,4]

print(list(map(lambda x: x % 2 == 0, nums)))
print(list(filter(lambda x: x % 2 == 0, nums)))
````

### `reduce()` Function:


## What is reduce() Function in Python?

The reduce() function in Python performs cumulative operations on iterables. It takes two main arguments: a function and an iterable. By applying the function cumulatively to the iterable’s elements, reduce() reduces them to a single value. This makes it particularly useful for tasks such as summing numbers or finding the product of elements in a list.

## How Does reduce() Work?

The reduce() function starts with the first two elements of an iterable, applies the function to them, then uses the result with the next element. This process continues until all elements are processed, resulting in a single cumulative value.

#### **Syntax and Parameters**

To use the reduce() function, import it from the functools module. The basic syntax is:

```python
from functools import reduce

result = reduce(function, iterable, [initializer])

```

**Explanation of Parameters:**

- **function**: The function to apply to the elements of the iterable. It must take two arguments.
- **iterable**: The iterable whose elements you want to reduce. It can be a list, tuple, or any other iterable.
- **initializer** (optional): The starting value. It is used as the first argument in the first function call if provided.


````python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

total = reduce(lambda x,y: x + y , numbers)

print(total)
````

````python

# Calculate product
product = reduce(lambda x, y: x * y, numbers)
print(product)
````

````python
maximum = reduce(lambda x, y: x if x > y else y, numbers)
print(maximum)
````

````python
# String concatenation
words = ['Python', 'is', 'awesome']
sentence = reduce(lambda x, y: x + ' ' + y, words)
print(sentence)
````
